```python
import numpy as np  # быстрые математические операции
import pandas as pd # для считывания данных
import matplotlib.pyplot as plt # для рисования графиков
```

```python
plt.plot([1, 2, 3, 4, 5], marker='*')
plt.show()
```

```python
plt.plot([1, 2, 3, 2, 1])
plt.plot([1, 2, 3, 4, 5])
plt.show()
```

```python
xs = range(10) # 0, 1, ..., 9
ys = [x**2 for x in xs] # 0, 1, 4, 9, 16, ..., 9*9
plt.plot(xs, ys, label='x^2')
plt.plot(xs, xs, label='x')
plt.legend() # если не вызвать, то легенда не появится
plt.show()
```

```python
xs = np.arange(0, 20, 0.1) # массив чисел от 0 до 20 с шагом 0.1
ys = np.sin(xs) # синус от каждого элемента массива
plt.plot(xs, ys)
plt.show()
```
